```r
new_data<-mutate(data1, video_category=as.factor(category))

cleaned<-select(new_data, -rank, -category, -Youtuber)


count_cat<-cleaned%>%
    group_by(video_category)%>%
    tally()
#note: this is to select the categories with the highest counts (we found that gaming/entertainment/music/people and blogs
#had the highest (all counts n>100), others were excluded) 

sorted<-arrange(count_cat, by_group=TRUE)


data2<-cleaned%>%
    filter(video_category %in% c("Gaming", "Music", "Enterntainment", "People and Blogs"))

data<-data2%>%
    filter(subscribers>0, video.views>0, video.count>0)%>%
    drop_na()
```

The goal of this analysis is to explore the factors which contribute to total channel views across 4 different categories of videos, Entertainment, Gaming, Music and Blogs and People. The response variable we will be exploring is the total channel views, and we will use different variables to estimate this. 

Note: we have choosen to exclude NA values, as well as values of 0 (as in channels with 0 videos, and thus 0 views), because these data points are outliers and will skew the data 


```r
#step 1: 
# Use forward selection algorithm to select variables (best variables for the model) 
variable_selection<-regsubsets(x=video.views~.,data=data, method="forward")
```
